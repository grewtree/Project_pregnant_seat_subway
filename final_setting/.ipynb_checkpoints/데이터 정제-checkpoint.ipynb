## 1. 기존 그대로 컬럼 이름만 변경
: date, time, line, start, stop, crowdedness, old_age_seat, is_empty, is_empty_count, age, badge, pregnant, heavy

```python
table = table.rename(columns = {
    '날짜': 'date',
    '기록 시간': 'time',
    '호선': 'line',
    '탑승역': 'start',
    '하차역': 'stop',
    '붐빔 정도': 'crowdedness',
    '노약자석 빈자리 여부': 'old_age_seat',
    '비어있었던 역 수': 'is_empty_count',
    '연령대': 'age',
    '뱃지 여부': 'badge',
    '임산부 여부': 'pregnant',
    '무거운 짐 여부': 'heavy',
})
```

## 2. 뱃지여부, 임산부여부, 무거운 짐 여부 O(있음 1), X(없음 0) 처리

```python
table['badge'].unique()
```

```python
table['badge'] = table['badge'].replace({'O': 1, 'X': 0})
```

```python
table['badge'].unique()
```

```python
table['pregnant'].unique()
```

```python
table['pregnant'] = table['pregnant'].replace({'O': 1, 'X': 0, '?': 0})
```

```python
table['pregnant'].unique()
```

```python
table['heavy'].unique()
```

```python
table['heavy'] = table['heavy'].replace({'O': 1, 'X': 0})
```

```python
table['heavy'].unique()
```

```python
table
```

## 3. 노약자석 빈자리 여부 0(빈자리 있음 O), 1(빈자리 없음) 처리

```python
table['old_age_seat'] = table['old_age_seat'].replace({'O': 0, 'X': 1})
```

```python
table
```

## 4. 성별 여자, 남자, 비었음으로 분리

```python
table['성별'].unique()
```

```python
table['is_empty'] = table['성별'].replace({'여': 0, '남': 0, '비었음': 1})
```

```python
table['female'] = table['성별'].replace({'여': 1, '남': 0, '비었음': 0})
```

```python
table['male'] = table['성별'].replace({'여': 0, '남': 1, '비었음': 0})
```

```python
table
```

## 5. 붐빔정도 분리

```python
table['crowdedness'].unique()
```

```python
table['E'] = table['crowdedness'].replace({'E': 1, 'N': 0, 'F': 0})
```

```python
table['N'] = table['crowdedness'].replace({'E': 0, 'N': 1, 'F': 0})
```

```python
table['F'] = table['crowdedness'].replace({'E': 0, 'N': 0, 'F': 1})
```
