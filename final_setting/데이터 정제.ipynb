## 6. 연령대 분리

child, teenage, youth, middle_age, old_age

```python
table['age'].unique()
```

```python
# nan 값 처리는 fillna() 함수 사용
```

```python
table['child'] = table['age'].replace({'중년': 0, '청년': 0, '노년': 0, '어린이': 1, '청소년': 0}).fillna(0).astype(int)
```

```python
table['teenage'] = table['age'].replace({'중년': 0, '청년': 0, '노년': 0, '어린이': 0, '청소년': 1}).fillna(0).astype(int)
```

```python
table['youth'] = table['age'].replace({'중년': 0, '청년': 1, '노년': 0, '어린이': 0, '청소년': 0}).fillna(0).astype(int)
```

```python
table['middle_age'] = table['age'].replace({'중년': 1, '청년': 0, '노년': 0, '어린이': 0, '청소년': 0}).fillna(0).astype(int)
```

```python
table['old_age'] = table['age'].replace({'중년': 0, '청년': 0, '노년': 1, '어린이': 0, '청소년': 0}).fillna(0).astype(int)
```
