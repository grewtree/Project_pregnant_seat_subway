```python
df['behavior'] = df['behavior'].replace(['핸드폰 게임','핸드폰(영상시청)'], '핸드폰')
df['behavior'] = df['behavior'].replace(['아이와 대화','친구와 대화'], '대화')
df['behavior'] = df['behavior'].replace(['X','미어캣','주변 살피기','묵주 돌리기','봉에 팔 걸침'], '멍')
df['behavior'] = df['behavior'].replace('공부', '필기')
df['behavior'] = df['behavior'].replace(['전단지 정독','책'], '독서')
df['behavior'] = df['behavior'].replace(['빵먹음','음료마심'], '먹기')
df['behavior'] = df['behavior'].replace(['부채질','손풍기쐬기','손풍기'], '바람쐬기')
df['behavior'] = df['behavior'].replace(['전화','핸드폰(전화통화)'], '통화')
```

```python
df['behavior'].unique()
```

```python
df['behavior'].value_counts()
```

```python
df['behavior'] = df['behavior'].replace('핸드폰 -> 독서','핸드폰, 독서')
df['behavior'] = df['behavior'].replace('핸드폰 -> 잠','핸드폰, 잠')
```

```python
df['behavior'].value_counts()
```

```python
df[['behavior1','behavior2','behavior3']] = df['behavior'].str.split(",",expand=True)
```

```python
df['behavior2'] = df['behavior2'].str.strip().replace('핸드폰(영상시청)', '핸드폰')
df['behavior2'] = df['behavior2'].replace(['미어캣','주변 살피기'], '멍')
df['behavior2'] = df['behavior2'].str.strip().replace('손풍기쐬기', '바람쐬기')
df['behavior2'].unique()
```

```python
def behavior_func(word):
    wordlist = []
    for i in range(len(df)):
        if word==df['behavior1'][i]:
           wordlist.append(1)
        else:
            if word==df['behavior2'][i]:
                wordlist.append(1)
            else:
                if word==df['behavior3'][i]:
                    wordlist.append(1)
                else:
                    wordlist.append(0)
    df[word] = pd.Series(wordlist)
    return
```

```python
behavior_func('핸드폰')
```

```python
df['behavior1'].value_counts()
```

```python
behavior_func('멍')
behavior_func('잠')
behavior_func('이어폰')
behavior_func('대화')
behavior_func('통화')
behavior_func('독서')
behavior_func('화장')
behavior_func('먹기')
behavior_func('필기')
behavior_func('바람쐬기')
```
