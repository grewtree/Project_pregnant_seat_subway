```python
table = pd.read_csv('data_pregnant.csv')
```

```python
table.columns
```

```python
table['count'] = 1
```

```python
table['sex'] = table['female'].apply(lambda x: "여" if x == 1 else "남" if x == 0 else '')
```

```python
table['sex'] = table['sex'].replace("",np.nan, regex=True)
```

```python
table.sex.unique()
```

```python
(plot_rows, plot_cols) = (2, 3)
size = 7

f, axes = plt.subplots(plot_rows, plot_cols, figsize=(plot_cols*size,plot_rows*size))
axes = np.concatenate( axes)

for (ax, col) in zip(axes,['is_empty', 'age', 'behavior1', 'heavy', 'cellphone', 'sex']):
    grouped_table = table.groupby(col).count()[['count']]
    grouped_table.plot(kind='pie', y='count', subplots=False, autopct='%1.1f%%', ax=ax, legend=False
                   , startangle=90,)
    ax.set_xlabel("["+col+"]")
    ax.set_ylabel('')
```

```python
def drawCommonPassenger(ds, topn=10, color=None):
    ds.groupby(['age', 'sex'], as_index=True)\
                .agg({'count':len})\
                .sort_values('count', ascending=False)[:topn]\
                .sort_values('count', ascending=True)\
                .plot(kind='barh', color=color, figsize=(6,int(topn/1.5)), title='임산부석에 앉는 가장 흔한 승객 유형 Top %d'  % topn )
    plt.ylabel("")
```

```python
drawCommonPassenger(table, topn=10)
```

```python
def drawCompareFeature(ds, feature1, feature2, title="", color=None):
    grouped_ds = ds.groupby([feature1, feature2], as_index=False)[['count']]\
            .count()\
            .pivot_table(index=feature2, columns=feature1, values='count', fill_value=0)\
    
    fig1, (ax1, ax2) = plt.subplots(ncols=2,  figsize=(15,5))
    plot = grouped_ds.plot(kind='bar', stacked=True,alpha=1, ax=ax1, color=color)
    ((grouped_ds.T/grouped_ds.sum(axis=1)).T).set_xticklabels(chart.get_xticklabels(), rotation=45).plot(kind='bar', stacked=True,alpha=1, ax=ax2, color=color)
    plt.tight_layout(pad=1.1).set_xticklabels(chart.get_xticklabels(), rotation=45)
    return plot
    
```

```python
chart = drawCompareFeature(table, 'sex', 'behavior1', "성별에 따른 행동비교")
```

```python
chart = drawCompareFeature(table, 'sex', 'age', "성별에 따른 나이 비교")
chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
```
